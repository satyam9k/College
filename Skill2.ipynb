### Data Visualization

```python
df = pd.read_csv('Marks.csv')
df
```

```python
df.drop(columns=['S. No', 'PRN Number',], axis=1)
```

### 1)Show the differnce in marks scored in each subjects

```python
df["DS101"].plot(figsize=(12,4.5),kind = "line",color = "red",label="DS101",linewidth = 1,alpha=1,grid=True,linestyle='-.')
df["DS102"].plot(kind = "line",color="green",label="DS102",linewidth=1,alpha=1,grid=True,linestyle="-")
df["DS103"].plot(kind = "line",color = "black",label = "DS103",linewidth=1,alpha=1,grid=True,linestyle=":")
df["DS104"].plot(kind = "line",color="blue",label="DS104",linewidth=1,alpha=1,grid=True,linestyle="-")
df["DS105"].plot(kind = "line",color = "yellow",label = "DS105",linewidth=1,alpha=1,grid=True,linestyle=":")
df["DS106"].plot(kind = "line",color="purple",label="DS106",linewidth=1,alpha=1,grid=True,linestyle="-")
plt.xlabel('Students',FontSize = 10,color = "purple")
plt.ylabel("Scores",FontSize = 10, color = "green")
plt.savefig("Graphic.png")
plt.show() 
```

### 2)Show Relation between marks scored in DS101 & DS102 by students.


```python
df.plot(kind = 'scatter',x="DS101",y="DS102",color="black",figsize=(17,8))
plt.xlabel("DS101",FontSize = 18)
plt.ylabel("DS102",FontSize = 18)
plt.title("Relation between marks scored",FontSize = 20)
plt.show()
```
